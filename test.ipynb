```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Data

```python
df = pd.read_csv("breast+cancer+wisconsin+diagnostic/wdbc.data", header=None)
```

```python
# Define column names
columns = ['ID', 'Diagnosis']

features = [
    'radius', 'texture', 'perimeter', 'area', 'smoothness',
    'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension'
]

# Add suffixes for mean, standard error, and worst
suffixes = ['mean', 'se', 'worst']

for suffix in suffixes:
    for feature in features:
        columns.append(f"{feature}_{suffix}")

# Assign to DataFrame
df.columns = columns
```

## Data Preprocessing

```python
mapping = {'M': 1, 'B':0}
df['Diagnosis'] = df['Diagnosis'].map(mapping)
```
